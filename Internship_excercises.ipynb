###Import libraries

```python
import numpy as np
import pandas as pd
import sklearn
```

### Data generating

```python
data = {
    "size": ["XL", "L", "M", "L", "M"],
    "color": ["red", "green", "blue", "green", "red"],
    "gender": ["female", "male", "male", "female", "female"],
    "price": [199.0, 89.0, 99.0, 129.0, 79.0],
    "weight": [500, 450, 300, 380, 410],
    "bought": ["yes", "no", "yes", "no", "yes"]
}

#Utworzenie z obiektu DataFrame na podstawie wy≈ºej zdafiniowanego obiektu
df_raw = pd.DataFrame(data = data)
df_raw
```

###Create a copy and describe the dataframe

```python
df = df_raw.copy()
df.info()
```

###Changing some of the object-type data to categorical data

```python
# Categorical data allow for easier visualisations
for col in ["size", "color", "gender", "bought"]:
  df[col] = df[col].astype("category")
df["weight"] = df["weight"].astype("float")

# Only shows data for columns with numeric values
df.describe()
df.describe(include=['category'])
```

###Using LabelEncoder




```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#le.fit(df['bought'])
#le.transform(df['bought'])
#le.fit_transform(df['bought']) executes fit and then transform
#You were doing that with mushrooms too so you could use one-hot encoding!
df['bought'] = le.fit_transform(df['bought'])
df
```

###Possible transform inversion

```python
# LabelEncoder should be only used for target data
df['bought'] = le.inverse_transform(df['bought'])
df
```

###Summary

```python
# Changing types
df_new = df_raw.copy()
for col in ["size", "color", "gender", "bought"]:
  df_new[col] = df_new[col].astype("category")
df_new["weight"] = df_new["weight"].astype("float")

# Encoding True as 1 and False as 0
le_new = LabelEncoder()
df_new['bought'] = le_new.fit_transform(df_new['bought'])

# Change categoricals to binary data with onehot encoding
df_new = pd.get_dummies(data=df_new, drop_first=True)

df_new[['size_M',	'size_XL',	'color_green',	'color_red', 'gender_male']] = df_new[['size_M',	'size_XL',	'color_green',	'color_red',	'gender_male']].apply(LabelEncoder().fit_transform)

# Standardization
scaler_new = StandardScaler()
df_new[['price', 'weight']] = scaler_new.fit_transform(df_new[['price', 'weight']])
df_new
```
