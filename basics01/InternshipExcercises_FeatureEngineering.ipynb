###Import libraries

```python
import numpy as np
import pandas as pd
import sklearn
```

###Load data

```python
def fetch_financial_data(company='AMZN'):
  ###This function fetches stock market quatations
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')
df_raw = fetch_financial_data()
df_raw.head()
```

###Creating data copy

```python
df = df_raw.copy()
df = df[:5]
df.info()
```

###Generationg new features

```python
# index = Date column
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

###Discretization

###Height

```python
df = pd.DataFrame(data={'height' : [175.,178.5,185.,191.,184.5,183.,168.]})
df
```

```python
df['height_cat'] = pd.cut(x=df['height'], bins=3)
df
```

```python
# Note that there are only 3 labels for 4 bin values
df['height_cat'] = pd.cut(x=df['height'], bins=(160, 175, 180,195), labels=['short', 'medium', 'tall',])
df
```

```python
pd.get_dummies(df, drop_first=True, prefix='height')
```

###App languages

```python
df = pd.DataFrame(data={'lang':[['PL', 'ENG'],['GER', 'ENG', 'PL', 'FRA'],['RUS']]})
df
```

```python
df['lang_number'] = df['lang'].apply(len)
df
```

```python
df['Aval.PL'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0)
```
